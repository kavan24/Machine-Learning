```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv('Melbourne_housing_FULL.csv')
```

```python
cols_to_use = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
```

```python
dataset=dataset[cols_to_use]
```

```python
dataset.head()
```

```python
dataset.isna().sum()
```

```python
cols_tofill_zero = ['Propertycount','Distance','Bedroom2','Bathroom','Car']
```

```python
dataset.isna().sum()
```

```python
dataset[cols_tofill_zero]=dataset[cols_tofill_zero].fillna(0)
```

```python
dataset.isna().sum()
```

```python
dataset['Landsize']=dataset['Landsize'].fillna(dataset.Landsize.mean())
```

```python
dataset['BuildingArea']=dataset['BuildingArea'].fillna(dataset.Landsize.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isna().sum()
```

```python
dataset.head()
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset.head()
```

```python
X=dataset.drop('Price',axis=1)
```

```python
y=dataset['Price']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg1=LinearRegression().fit(train_x,train_y)
```

```python
reg1.score(test_x,test_y)
```

```python
reg1.score(train_x,train_y)
```

```python
from sklearn import linear_model
```

```python
lasso_reg=linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(train_x,train_y)
```

```python
lasso_reg.score(test_x,test_y)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_reg = Ridge(alpha=50,max_iter=100,tol=0.1)
```

```python
ridge_reg.fit(train_x,train_y)
```

```python
ridge_reg.score(test_x,test_y)
```
